```python
def zscore_normalize(X):
    mu=np.mean(X,axis=0)
    sig=np.std(X,axis=0)
    X_norm=(X-mu)/sig
    return X_norm
```

```python
def sigmoid(z):
    y=1/(1+np.exp(-z))
    return y
```

```python
def cost_regularization(X,Y,w,b,lamda=0):
    m,n=X.shape
    cost=0.
    for i in range (m):
        z_i=np.dot(X[i],w)+b
        f_wb_i=sigmoid(z_i)
        #cost+= -Y[i]*np.log(f_wb_i)-(1-Y[i])*np.log(1-f_wb_i)
        if (Y[i]==0):
            cost=cost+np.log(1-f_wb_i)
        else:
            cost=cost+np.log(f_wb_i)
        
    cost=cost/m
    reg_cost=0;
    for j in range(n):
        reg_cost+= w[j]**2
    
    reg_cost=(lamda*reg_cost)/(2*m)
    total_cost=reg_cost+cost
    return total_cost
```

```python
def gradient_regularization(X,Y,w,b,lamda=0):
    m,n=X.shape
    dj_dw=np.zeros((n,))
    dj_db=0.
    
    for i in range(m):
        f_wb_i=sigmoid(np.dot(X[i],w)+b)
        err_i=f_wb_i - Y[i]
        for j in range(n):
            dj_dw[j]+=err_i*(X[i,j])
        dj_db+=err_i
            
    dj_dw=dj_dw/m
    dj_db=dj_db/m
    
    for i in range(n):
        dj_dw[i]+=(lamda/m)*(w[i])
        
    return dj_db,dj_dw
```

```python
def gradient_descent(X,Y,w_in,b_in,alpha,num_iters,lamda=0):
    m=len(X)
    J_history=[]
    w_history=[]
    b_history=[]
    index=[]
    
    for i in range(num_iters):
        dj_db,dj_dw=gradient_regularization(X,Y,w_in,b_in,lamda)
        
        w_in=w_in-alpha*dj_dw
        b_in=b_in-alpha*dj_db
        
        if i%100 == 0:
            cost=cost_regularization(X,Y,w_in,b_in,lamda)
            J_history.append(np.abs(cost))
            w_history.append(w_in)
            b_history.append(b_in)
            index.append(i)
        
    return w_in,b_in,J_history,w_history,index,b_history
        
    
```

```python
def plot_(J_history,index,lamda=0):
    plt.plot(index, J_history)
    #plt.xticks(index, rotation ='vertical')
    plt.xlabel("Number of Iterarion")
    plt.ylabel("Cost")
    plt.title('Regularization with lambda_ = {}'.format(lamda))
    plt.show()
```

```python
def predict(w,b,X):
    m,n=X.shape
    y_predic = np.zeros((m,))
    for i in range(m):
        z=sigmoid(np.dot(X[i],w)+b)
        if z <= 0.5:
            y_predic[i]=0
        else:
            y_predic[i]=1
    return y_predic
```

```python
def logidtic_regression(x_train,y_train,x_test,y_test,alpha,num_iters,lamda):
    b_in=0.
    n=x_train.shape[1]
    w_in=np.zeros((n,))
    w,b,J_history,w_history,index,b_history=gradient_descent(x_train,y_train,w_in,b_in,alpha,num_iters,lamda)
    y_pre_test=predict(w,b,x_test)
    y_pre_train=predict(w,b,x_train)
    
    #print accuracy
    print("train accuracy : {} %".format(100-np.mean(np.abs(y_pre_train-y_train))*100))
    print("test accuracy : {} %".format(100-np.mean(np.abs(y_pre_test-y_test))*100))
    
    return J_history,index,w_history,b_history
```

```python
def logidtic_regression1(x_train,y_train,alpha,num_iters,lamda):
    b_in=0.
    n=x_train.shape[1]
    w_in=np.zeros((n,))
    w,b,J_history,w_history,index,b_history=gradient_descent(x_train,y_train,w_in,b_in,alpha,num_iters,lamda)
    y_pre_train=predict(w,b,x_train)
    
    #print accuracy
    print("train accuracy : {} %".format(100-np.mean(np.abs(y_pre_train-y_train))*100))
    
    return J_history,index,w_history,b_history
```
